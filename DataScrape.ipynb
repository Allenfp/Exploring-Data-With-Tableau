```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
from time import sleep
```

```python
browser = Browser('chrome', headless=False)
url = 'https://s3.amazonaws.com/tripdata/index.html'
browser.visit(url)
sleep(2.0)
html = browser.html
soup = bs(html, 'lxml')
browser.quit()
```

```python
links = []
for link in soup.find_all('a'):
    links.append(link.get('href'))
dl_links = links[:-1]
dl_links
```

```python
import wget
from zipfile import *
import os

for link in dl_links:
    if link != link.startswith('https://s3.amazonaws.com/tripdata/JC'):
        wget.download(link, 'unzipme.zip')
        zip = ZipFile('unzipme.zip')
        zip.extractall()
        os.remove('unzipme.zip')
```
